## 8.图片读写及显示

```python
import numpy as np
from matplotlib import pyplot as plt
import cv2 as cv
```

```python
img = cv.imread("pic/bear500x333.jpg")
cv.imwrite("pic/test/bear.jpg",img)
```

```python
cv.imshow("bear",img)
#cv.waitKey(0)#一直等待
#cv.waitKey(100)#100ms
cv.destroyAllWindows()
```

```python
plt.imshow(img)
plt.show()
```

```python
#RGB  matplotlib -> BGR opencv
img2 = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img2)
plt.show()
```

```python
#灰度图
img3 = cv.imread("pic/bear500x333.jpg",0)
cv.imshow("bear",img3)
cv.waitKey(1)
cv.destroyAllWindows()
```

```python
plt.imshow(img3,cmap="gray")#plt中将灰度图以热力图的形式，要加上cmap="gray"
plt.show()
```

```python
img3.ndim #灰度图为2维
```

```python
img2.ndim #彩色图为3维
```

```python
#读图片函数封装
def ImageShow(img):
    if(img.ndim==3):
        img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
        plt.imshow(img)
    else:
        plt.imshow(img,cmap="gray")
    plt.show()      
```

```python
apple = cv.imread("pic/apple.jpg")
ImageShow(apple)
```

```python
apple_gray = cv.imread("pic/apple.jpg",0)
ImageShow(apple_gray)
```

## 9.简单绘图

```python
#plot 
```

```python
x = np.arange(2,20)
y = 2*x+np.random.randint(0,10,18)
plt.plot(x,y,"*-",color="r")
plt.show()
```

```python
x = np.linspace(0,1,100)
y = np.power(x,0.5)
y2 = x
y3 = np.power(x,1.5)
#plt.plot(x,y,x,y2,x,y3)
plt.plot(x,y,label="0.5")
plt.plot(x,y2,label="1")
plt.plot(x,y3,label="1.5")
plt.xlabel("x")
plt.ylabel("s")
plt.legend()
plt.grid(alpha=0.5)
plt.xlim([0,1])
plt.ylim([0,1])
plt.show()
```
