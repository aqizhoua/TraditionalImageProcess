## 8.图片读写及显示

```python
import cv2 as cv
```

```python
img = cv.imread("pic/apple.jpg")
cv.imwrite("images/apple.jpg",img)
```

```python
cv.imshow("apple",img)
cv.waitKey(100)
cv.destroyAllWindows()
```

```python
img2 = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img2)

plt.show()
```

```python
#灰度图
img = cv.imread("images/apple.jpg",0)
img2 = cv.cvtColor(img,cv.COLOR_BGR2RGB)#方法1
plt.imshow(img2)
plt.show()
```

```python
plt.imshow(img,cmap="gray")
plt.show()
```

```python
img.ndim
```

```python
img = cv.imread("images/apple.jpg",0)
img.ndim
```

```python
def show(img):
    if img.ndim==2:
        plt.imshow(img,cmap="gray")
    elif img.ndim==3:
        img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
        plt.imshow(img)
    plt.show()
```

```python
show(img)
```

## 9.简单绘图

```python
#plot
x=np.arange(10)
y=x+np.random.randint(1,10,10)
plt.plot(x,y,"*-",color="red")
plt.show()
```

```python
x=np.linspace(0,1,100)
y1=np.power(x,0.5)
y2=np.power(x,1.0)
y3=np.power(x,2)
plt.plot(x,y1,label="x^0.5")
plt.plot(x,y2,label="x")
plt.plot(x,y3,label="x^2")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.xlim([0,1])
plt.ylim([0,1])
plt.show()
```

```python
#hist
a=np.random.randint(0,100,1000)
plt.hist(a,rwidth=0.9,color="r")
plt.show()
```

```python
bins=np.arange(-0.5,101,1.0)
plt.hist(a,bins)
plt.show()
```
