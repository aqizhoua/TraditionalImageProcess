```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(img):
    if img.ndim == 2:
        plt.imshow(img, cmap='gray')
        plt.show()
    else:
        plt.imshow(img)
        plt.show()
```

```python
img = cv2.imread('./pic/cake1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
show(img)
```

```python
show(img)
```

```python
img2 = img.copy()

for i in range(270, 335):
    for j in range(310, 500):
        for c in range(3):
            if img2[i,j,c] < 190:
                img2[i,j,c] = img2[i,j,c] + np.random.randint(40, 65)
```
