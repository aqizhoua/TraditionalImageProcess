```python
#封装为函数
def show(img):
    if img.ndim==2:
        plt.imshow(img,cmap="gray")
    else:
        plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
    plt.show()
```

```python
show(img)
```

```python
show(img3)
```

# 8.简单绘图

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(2,20)
y = 2*x+np.random.randint(5,20,18)
print(np.random.randint(5,20,18))
```

```python
plt.plot(x,y,"*-",color="r")
plt.show()
```

```python
#同一张图中显示多条曲线
x = np.linspace(0,1,100)
y1 = np.power(x,0.5)
y2 = x
y3 = np.power(x,1.5)

plt.plot(x,y1,x,y2,x,y3)
plt.show()
```

```python
#显示图例
plt.plot(x,y1,label="0.5")
plt.plot(x,y2,label="1")
plt.plot(x,y3,label="1.5")
plt.xlabel("r")
plt.ylabel("s")
plt.grid(alpha=0.5)
plt.legend()
plt.xlim([0,1]) #限制x,y坐标范围
plt.ylim([0,1])
plt.show()
```

```python
#直方图
a = np.random.randint(0,101,1000)

plt.hist(a,rwidth=0.9,color="orange")
plt.show()
```
